```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import streamlit as st
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('matches.csv')
```

```python
df.shape
```

```python
df.head()
```

```python
df['match_type'].unique()
```

```python
df_drop_column_list=['date','venue','date','toss_winner','toss_decision','umpire1','umpire2','method','target_overs','result']
df.drop(df_drop_column_list,axis=1,inplace=True)
```

```python
df['season']=df['season'].str.split('/').str[0].astype(int)
```

```python
df.dropna(inplace=True)
```

```python
df['result_margin']=df['result_margin'].astype('int64')
df['target_runs']=df['target_runs'].astype('int64')
```

```python
df.head()
```

```python
df['season'].unique()
```

```python
df['id'] = range(1, len(df) + 1)
```

```python
df[df['season']==2007]['id'].unique()
```

```python
plt.figure(figsize=(15,5))
sns.countplot(data=df,x='season')
plt.show()
```

```python
df.head()
```

```python
plt.figure(figsize=(15,5))
winner_counts = df['winner'].value_counts()
sns.countplot(data=df,x='winner',order=winner_counts.index)
plt.xticks(rotation=90)
plt.show()
```

```python
df['match_type'].unique()
```

```python
temp=df[df['match_type']=='Final']
sns.countplot(data=temp,x='winner')
plt.xticks(rotation=90)
```

```python
team_name_mapping = {
    'Delhi Daredevils': 'Delhi Capitals',
    'Kings XI Punjab': 'Punjab Kings',
    'Rising Pune Supergiants': 'Rising Pune Supergiant',
    'Royal Challengers Bengaluru': 'Royal Challengers Bangalore',
}

team_columns = ['winner', 'team1', 'team2']

# Replace the team names in the 'winner' column
df[team_columns] = df[team_columns].replace(team_name_mapping)
```

```python
df.to_csv("clean_dataset.csv",index=False)
```
